# Exploratory Data Analysis (EDA)

### 1. Churn Distrubution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='churn', data=df)
plt.title('Churn Distribution')
plt.xlabel('Churn (0: No, 1: Yes)')
plt.ylabel('Count')
plt.savefig('churn_distribution.png')
plt.show()
```

```python
### 1. Churn Distribution

The graph in cell index 11 is a bar plot that visualizes the distribution of churn in the dataset. It uses the `sns.countplot()` function to display the count of customers who churned (`churn = 1`) versus those who did not churn (`churn = 0`).

Key Points:
- The x-axis represents the churn status (0 for no churn, 1 for churn).
- The y-axis represents the count of customers.
- The plot highlights the imbalance in the dataset, where the majority of customers did not churn.
- This visualization provides a quick overview of the churn distribution, which is crucial for understanding the dataset's class imbalance.
- The imbalance suggests the need for techniques like resampling or weighted metrics during model training to handle the skewed data effectively.
```

### 2. Age Distribution by Churn

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='churn', multiple='stack', bins=30)
plt.title('Age Distribution by Churn')
plt.xlabel('Age')
plt.ylabel('Count')
plt.savefig('age_churn_distribution.png')
plt.show()
```

```python
### 2. Age Distribution by Churn

The graph in cell index 14 is a histogram that visualizes the distribution of customer ages, segmented by their churn status. It uses the `sns.histplot()` function with the `hue` parameter set to `churn`, allowing for a stacked visualization of churned (`churn = 1`) and non-churned (`churn = 0`) customers.

Key Points:
- The x-axis represents the age of customers.
- The y-axis represents the count of customers.
- The `hue` parameter differentiates between churned and non-churned customers.
- The plot provides insights into the age groups that are more likely to churn or stay, helping identify patterns in customer behavior based on age.
- For example, younger or older age groups may show higher churn rates compared to middle-aged customers.
```

### 3. Telecom Partner vs Churn

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='telecom_partner', hue='churn', data=df)
plt.title('Churn by Telecom Partner')
plt.xlabel('Telecom Partner')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('telecom_churn.png')
plt.show()
```

```python
### 3. Telecom Partner vs Churn

The graph in cell index 16 is a bar plot that visualizes the relationship between telecom partners and churn status. It uses the `sns.countplot()` function to display the count of customers for each telecom partner, segmented by churn status (`churn = 1` for churned and `churn = 0` for non-churned customers).

Key Points:
- The x-axis represents the telecom partners (e.g., Reliance Jio, Vodafone, BSNL, Airtel).
- The y-axis represents the count of customers.
- The `hue` parameter differentiates between churned and non-churned customers.
- The plot highlights the distribution of churn across different telecom partners, helping identify which partners have higher or lower churn rates.
- This visualization can assist in understanding customer retention trends for each telecom partner.
```

### 4. Correlation Heatmap

```python
numeric_cols = ['age', 'num_dependents', 'estimated_salary', 'calls_made', 'sms_sent', 'data_used', 'churn']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
### 4. Correlation Heatmap

The graph in cell index 18 is a heatmap that visualizes the correlation between numeric variables in the dataset. It uses the `sns.heatmap()` function to display the pairwise correlation coefficients.

Key Points:
- The heatmap shows the strength and direction of relationships between numeric variables.
- The correlation values range from -1 to 1:
    - A value close to 1 indicates a strong positive correlation.
    - A value close to -1 indicates a strong negative correlation.
    - A value close to 0 indicates no correlation.
- The `annot=True` parameter displays the correlation values on the heatmap.
- The `cmap='coolwarm'` parameter provides a color gradient, where warm colors (red) indicate positive correlations and cool colors (blue) indicate negative correlations.
- This plot helps identify potential relationships between variables, such as the correlation between `data_used` and `calls_made`, or `estimated_salary` and `churn`.
```

### 5. Data Usage vs Estimated Salary by Churn

```python
### 5. Data Usage vs Estimated Salary by Churn

The graph in cell index 20 is a scatter plot that visualizes the relationship between estimated salary and data usage, segmented by churn status. It uses the `sns.scatterplot()` function with the `hue` parameter set to `churn`, allowing for differentiation between churned (`churn = 1`) and non-churned (`churn = 0`) customers.

Key Points:
- The x-axis represents the estimated salary of customers.
- The y-axis represents the data usage (in MB) by customers.
- The `hue` parameter is used to color-code the points based on churn status.
- The `size` parameter emphasizes the churn status visually.
- This plot helps identify patterns or clusters in data usage and salary that may correlate with churn behavior.
```

### 8. Time Since Registration vs Churn

```python
# Convert 'date_of_registration' to datetime
df['date_of_registration'] = pd.to_datetime(df['date_of_registration'])

# Calculate days since registration
df['days_since_registration'] = (pd.Timestamp('2025-05-02') - df['date_of_registration']).dt.days

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='days_since_registration', hue='churn', multiple='stack', bins=30)
plt.title('Days Since Registration by Churn')
plt.xlabel('Days Since Registration')
plt.ylabel('Count')
plt.savefig('registration_churn.png')
plt.show()
```

```python
### 8. Time Since Registration vs Churn

The graph in cell index 27 is a histogram that visualizes the distribution of the number of days since registration for customers, segmented by their churn status. It uses the `sns.histplot()` function with the `hue` parameter set to `churn`, allowing for a stacked visualization of churned (`churn = 1`) and non-churned (`churn = 0`) customers.

Key Points:
- The x-axis represents the number of days since registration.
- The y-axis represents the count of customers.
- The `hue` parameter differentiates between churned and non-churned customers.
- This plot provides insights into how the duration of customer registration correlates with churn behavior.
```
